LOADING DATA

```python
import pandas as pd
```

```python
fake = pd.read_csv("../data/Fake.csv")
real = pd.read_csv("../data/True.csv")
```

```python
fake['label'] = 0
real['label'] = 1
```

CONCATING DATA

```python
df = pd.concat([fake, real])
df = df.sample(frac=1).reset_index(drop=True)

df.head()
```

```python
df['content'] = df['title'] + " " + df['text']
```

```python
df.head()
```

```python
# Keep only 'text', 'title', and 'label' columns, drop all others
df = df[['content', 'label']]

# Verify the result
print(f"DataFrame shape after keeping only essential columns: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
df.head()
```
